## Zomato dataset Exploratory data analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("zomato.csv", encoding = 'latin-1')
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

### Missing value

```python
df.isnull().sum()
```

```python
country_code = pd.read_excel("Country-Code.xlsx")
country_code
```

```python
final_df = pd.merge(df,country_code, on = 'Country Code', how = "left")
final_df.head()
```

```python
country_names = final_df['Country'].value_counts().index
```

```python
country_val = final_df['Country'].value_counts().values
```

```python
## Pie chart of top3 country
plt.pie(country_val[:3],labels = country_names[:3],autopct="%1.2f%%")
```

### zomato maximum records or transcation are from india and it perform very low in usa or uk ie insignificant in normal case

```python
final_df.columns
```

```python
rating = final_df.groupby(["Aggregate rating",'Rating color','Rating text']).size().reset_index().rename(columns = {0:'Rating count'})
```

###Observation: rating 4.5 to 4.9 excellent, 4.0 to 4.5 very good , 3.5 to 3.9 rating is good, 3.0 to 3.4 is average. 

```python
rating.head()
```

```python
plt.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y = "Rating count",hue = "Rating color", data=rating, palette=['white','Red','Orange','yellow','green','darkgreen'] )
```
